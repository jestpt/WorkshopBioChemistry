```python
# Exercise 1
# Display with the sns.distplot() function the distribution of first sexual intercourse

%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('../data/cervical_cancer.csv')

plt.figure()
sns.distplot(dataset["First sexual intercourse"].dropna())
```

```python
# Exercise 2
# Display with the sns.barplot() function the relationship between STDs and smoking
# in risk of developing cervical cancer or not

%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('../data/cervical_cancer.csv')

plt.figure()
plt.title('Relationship between STDs and smoking')
sns.set(color_codes=True)
sns.barplot(x = 'STDs', y = 'Smokes', hue = 'Dx:Cancer', data = dataset);
```
