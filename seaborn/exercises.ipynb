```python
# Exercise 1
# Display with the sns.distplot() function the distribution of first sexual intercourse

%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('../data/cervical_cancer.csv')

plt.figure()
sns.distplot(dataset["First sexual intercourse"].dropna())
```

```python
# Exercise 2
# Choose two variables you think make the most sense to correlate, and visualize their relationship

%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('../data/cervical_cancer.csv')

plt.figure()
plt.title('Relationship between STDs and smoking')
sns.set(color_codes=True)
sns.boxplot(x = 'STDs', y = 'Age', hue = 'Dx:Cancer', data = dataset);
```
