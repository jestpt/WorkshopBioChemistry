```python
#######################################################################
############ Statistical Data Visualization with Seaborn ##############
#######################################################################

# More info at: https://seaborn.pydata.org/
# For all the possible plots: https://seaborn.pydata.org/tutorial.html
# Dataset: https://www.kaggle.com/loveall/cervical-cancer-risk-classification/data
```

```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Dataset
dataset = pd.read_csv('../data/cervical_cancer.csv')

list(dataset)
#print(dataset['Hormonal Contraceptives (years)'])
```

```python
# Visualizing distribution of Age

# Distplot
plt.figure()  # To create a figure
plt.title('Cervical Cancer Distribution by Age')  
sns.distplot(dataset['Age'].loc[dataset['Dx:Cancer'] == 1], color = 'r', bins = 10)
sns.distplot(dataset['Age'].loc[dataset['Dx:Cancer'] == 0], color = 'b', bins = 20)
plt.legend(['Affected by cancer', 'Control'])
```

```python
plt.figure()
plt.title('Distribution of Number of Sexual Partners')
desired_array = dataset['Number of sexual partners'].dropna()
sns.distplot(desired_array, bins = 70)
plt.xlim((0,8))  # We can control the range 
plt.xlabel('Number of sexual partners')
```
